```python
'''
3. ¿Cuántos tipos diferentes de pokémon hay? (Type 1)
'''
tipos_pok = df['Type 1'].nunique()

print("Hay", tipos_pok, "tipos diferentes de pokémon")
```

```python
'''
2. ¿Cuántos pokémon hay de cada tipo? (Type 1)
'''
cuantos_pk = df['Type 1'].value_counts(ascending=False)
print(cuantos_pk)

'''
2.1 ¿cual es el tipo de pokémon con más pokémon? (Type 1)

'''
print()

cuantos_pk_max = df['Type 1'].value_counts().idxmax()

print("El tipo de pokémon con más pokémon es", cuantos_pk_max )

```

```python
'''
3. ¿Cuál es el pokémon más rápido?
'''
max_vel = df['Speed'].max()

max_pok = df[df['Speed'] == max_vel]["Name"]

print("El pokémon más rápido es", max_pok )
```

```python
'''
4. ¿Cuántos pokemon tiene una defensa superior a 100?
'''

defensa_100 = df['Defense'] > 100
pok_100 = len(df[defensa_100])

print(pok_100, "pokemon tiene una defensa superior a 100" )
```

```python
'''
9. Calcula el promedio de stats de los pokemon Legendary (HP, Attack, Defense, Sp. Atk, Sp. Def, Speed) y los no Legendary
'''
legendary = df[(df['Legendary'] == True)]
no_legendary = df[(df['Legendary'] == False)]

mean_legendary = legendary[["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]].mean().round(2)
mean_no_legendary = no_legendary[["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]].mean().round(2)

print("Promedio Legendary\n", mean_legendary)
print("Promedio No Legendary\n", mean_no_legendary)
```

```python
from functools import reduce

'''
13. Dada una lista de Artículos con sus precio. Define las siguientes funciones:
Puedes definir más funciones si lo consideras necesario.
'''
articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']


# A. Una función que calcule el precio total de la compra

precio_total_compra = reduce(lambda x, y: x + articulos[y] if y in articulos else 0, compra, 0) 

print('Total compra:', precio_total_compra)

# B. Una función que calcule el precio total de la compra con un descuento del 10%

descuento_10 = lambda precio_total_compra:  precio_total_compra - (precio_total_compra*0.01)  

print('Total con descuento 10%:' , descuento_10(precio_total_compra))

# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€

if precio_total_compra > 100:
    descuento_10 = lambda precio_total_compra:  precio_total_compra - (precio_total_compra*0.01)  
    print('Total con descuento 10% si mayor a 100€:' , descuento_10(precio_total_compra))
else:
    print('Total:' , precio_total_compra)

# D. Una función que calcule el precio total aplicando el IVA (21%)

IVA = lambda precio_total_compra: precio_total_compra*1.21  
print('Total con IVA:' , precio_total_compra)

# E.lista los artículos cuyo precio es superior a 20€

sup_20 = list(filter(lambda item: True if item in articulos and articulos[item]>20 else False,compra)) ; 

print('Productos superiores a 20€:', sup_20)
```
